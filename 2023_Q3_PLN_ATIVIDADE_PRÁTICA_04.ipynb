```
#tecnica de extração de palavras chaves. CÉLULA 1

import requests
import json

def openai_complete(prompt):
    # Defina sua chave de API da OpenAI
    api_key = 'COLOQUE AQUI SUA API KEY' #conforme orientado em aula, removemos nossa api key

    # Defina os headers da requisição
    headers = {
        'Content-Type': 'application/json',
        'Authorization': f'Bearer {api_key}'
    }

    # Defina o payload da requisição
    data = {
        'prompt': prompt,
        'max_tokens': 100,  # Defina o número máximo de tokens na resposta gerada
        'temperature': 0.7,  # Defina a "temperatura" para controlar a aleatoriedade das respostas (0.2 a 1.0)
        'n': 1  # Defina o número de respostas desejadas
    }

    # Faça a requisição POST para o endpoint da API da OpenAI
    response = requests.post('https://api.openai.com/v1/chat/completions', headers=headers, json=data)

    # Obtenha e retorne a resposta JSON
    return response.json()
```

```
#tecnica de extração de palavras chaves. CÉLULA 2
import requests
from bs4 import BeautifulSoup

def obter_texto_do_link(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        response.encoding = 'utf-8'
        # Parseia o conteúdo HTML usando BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        # Extrai somente o texto visível
        texto = soup.get_text(separator='\n', strip=True)
        return texto
    except requests.exceptions.RequestException as e:
        print(f"Ocorreu um erro ao acessar o link: {e}")
        return None
```

```
#tecnica de extração de palavras chaves. CÉLULA 3
import requests
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from bs4 import BeautifulSoup

def obter_palavras_chave_texto(texto):
    # Tokenização das palavras do texto
    tokens = word_tokenize(texto)

    # Remoção das stopwords (palavras irrelevantes)
    stop_words = set(stopwords.words('portuguese'))
    tokens_sem_stopwords = [token for token in tokens if token.lower() not in stop_words and token.isalpha()]

    # Frequência das palavras
    fdist = FreqDist(tokens_sem_stopwords)

    # Obtenção das palavras-chave (as mais frequentes)
    palavras_chave = fdist.most_common(10)  # Altere o número '10' para a quantidade desejada

    return palavras_chave
```

```
#tecnica de analise de sentimentos. CELULA 1

import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize

def analisar_sentimentos_link(link):
    # Obter o conteúdo HTML da página
    response = requests.get(link)
    html = response.text

    # Extrair o texto do HTML usando BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    texto = soup.get_text()

    # Tokenização das palavras do texto
    tokens = word_tokenize(texto)

    # Carregamento de dicionários de palavras associadas a sentimentos
    nltk.download('opinion_lexicon')
    from nltk.corpus import opinion_lexicon

    # Contagem das palavras positivas e negativas
    palavras_positivas = 0
    palavras_negativas = 0

    for token in tokens:
        if token.lower() in opinion_lexicon.positive():
            palavras_positivas += 1
        elif token.lower() in opinion_lexicon.negative():
            palavras_negativas += 1

    # Determinação da polaridade com base nas contagens
    if palavras_positivas > palavras_negativas:
        polaridade = "positivo"
    elif palavras_positivas < palavras_negativas:
        polaridade = "negativo"
    else:
        polaridade = "neutro"

    return polaridade

link = "https://brasileiraspln.com/livro-pln/1a-edicao/parte5/cap8/cap8.html"  # Substitua pela URL desejada

resultado = analisar_sentimentos_link(link)
print("Polaridade emocional:", resultado)
```

```
#tecnica de correção gramatical. CELULA 2

from language_tool_python import LanguageTool

def corrigir_gramatica(texto):
    tool = LanguageTool('pt')
    texto_corrigido = tool.correct(texto)

    matches = tool.check(texto)
    palavras_corrigidas = []
    for match in matches:
        if match.replacements:
            palavra_original = texto[match.offset:match.offset + match.errorLength]
            palavra_corrigida = match.replacements[0]
            # Ignora a correção se a palavra original for um nome próprio desejado
            texto_corrigido = texto_corrigido.replace('ítalo', 'Ítalo')
            if palavra_original.lower() == "ítalo":
                palavras_corrigidas.append((palavra_original, palavra_original))
            else:
                palavras_corrigidas.append((palavra_original, palavra_corrigida))

    return texto_corrigido, palavras_corrigidas

# Seu código para obter o texto do link
url = "https://brasileiraspln.com/livro-pln/1a-edicao/parte5/cap8/cap8.html"
texto = obter_texto_do_link(url)

if texto is not None:
    # Aplicando a correção gramatical ao texto
    texto_corrigido, palavras_corrigidas = corrigir_gramatica(texto)

    # Exibir o texto corrigido
    print("Texto corrigido:")
    print(texto_corrigido)

    # Exibir palavras originais e corrigidas lado a lado
    print("\nPalavra antes da correção | Palavra depois da correção:")
    for palavra_original, palavra_corrigida in palavras_corrigidas:
        print(f"{palavra_original} | {palavra_corrigida}")
        print("\n")
else:
    print("Não foi possível obter o texto do link.")
```
